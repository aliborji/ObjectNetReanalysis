```python
import os
import json
import glob
```

```python
with open("../mappings/folder_to_objectnet_label.json") as f:
    folders = json.load(f)
```

```python
count = 1
for f in folders:
    txtfile = '../' + f + '.txt'
    if not os.path.exists(txtfile):
        continue
        
    with open(txtfile) as ff:
        boxes = [line.strip() for line in ff.readlines() if len(line)>10] # excluding the last empty line    
    files = set([x.split(' ')[0] for x in boxes])    
        
    imgs = glob.glob('../images/' + f + '/*.png')    
    ims = [os.path.basename(x) for x in imgs]

    print(f"{count}, {txtfile}, {len(ims)}, lines = {len(boxes)}")
  
    assert len(imgs) == len(boxes), "Oh no! does not match!"

    if len(ims) > len(boxes):
        print(set(ims) - set(files))
        
    count += 1    
        
    
```
